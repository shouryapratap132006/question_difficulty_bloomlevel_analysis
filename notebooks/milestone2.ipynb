# 🎓 Milestone 2: Agentic AI Assessment Design Assistant

This notebook provides a purely programmatic environment to test and view the output of the **Milestone 2 LangGraph Agent**. It securely connects the ML predictions from Milestone 1 directly into the LLM logic flow without needing to launch Streamlit.

```python
import os
import json
from dotenv import load_dotenv
from logistic_regression_deployment import BloomModelDeployer
from graph import agent_graph

load_dotenv(override=True)
print('Environment Loaded. Groq API Key Configured:', bool(os.getenv('GROQ_API_KEY')))
```

```python
# 1. Load the Milestone 1 ML Model
worker = BloomModelDeployer()
worker.load_models()

# Sample Question Data
sample_data = {
    'question_text': 'Explain the process of photosynthesis in your own words.',
    'subject': 'Science',
    'topic': 'Biology',
    'avg_score': 0.70,
    'correct_percentage': 70.0,
    'num_students_attempted': 100,
    'num_students_correct': 70,
    'time_taken_minutes': 2.0
}

# Run ML Prediction
ml_results = worker.predict(sample_data)
print('\nML Predictions:', ml_results)
```

```python
# 2. Run the LangGraph Agentic Assistant (Milestone 2)
initial_state = {
    "question_data": sample_data,
    "ml_results": ml_results
}

print("Initiating LangGraph Pipeline...")
final_state = agent_graph.invoke(initial_state)
report = final_state.get('final_report')

print('Agent Execution Complete!')
```

```python
# 3. Display the Final Report
for key, value in report.items():
    print(f"{key.upper()}")
    print(value)
    print("-" * 50)
```
